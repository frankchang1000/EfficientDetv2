```python
# take an input mp4 file and split it into frames
# save the frames to disk

import cv2
import os

# Read the video from specified path
cam = cv2.VideoCapture("D:\coding\isef2023\data\downloads/treadmillv2/val/val.mp4")

try:
        # creating a folder named newdata
        if not os.path.exists('newdata'):
            os.makedirs('newdata')

# if not created then raise error
except OSError:
    print ('Error: Creating directory of data')
```

```python
# read all frames but only save every 10th frame
count = 22000
success, frame = cam.read()
while success:
    # if video is still left continue creating images

    count += 1

    if count % 10 == 0:
        name = 'D:\coding\isef2023\data\downloads/treadmillv2/val/' + str(count) + '.jpeg'
        print ('Creating...' + name)

        # writing the extracted images
        cv2.imwrite(name, frame)


    success, frame = cam.read()

# Release all space and windows once done
cam.release()
```

```python
import os
import tensorflow as tf
import xml.etree.ElementTree as ET

import dataset as dataset

def find_bad(path_to_folder, images_folder="Images", labels_folder="Labels"):
    file_names = os.listdir(os.path.join(path_to_folder, images_folder))
    bad_file = False
    bad_labels = []

    for x in file_names:
        bad_file = False
        # get the base name of the file
        base_name = os.path.splitext(x)[0]
        # get the path to the xml file

        file_name = os.path.join(path_to_folder, labels_folder, base_name + ".xml")
        root = ET.parse(file_name).getroot()

        boxes = root.findall("object")
        box = []
        labels = []

        for b in boxes:
            bb = b.find("bndbox")
            if int(bb.findtext("xmin")) >= int(bb.findtext("xmax")):
                print("Bad xmin: {}".format(bb.findtext("xmin")))
                print("Bad xmax: {}".format(bb.findtext("xmax")))
                bad_file = True
            if int(bb.findtext("ymin")) >= int(bb.findtext("ymax")):
                print("Bad ymin: {}".format(bb.findtext("ymin")))
                print("Bad ymax: {}".format(bb.findtext("ymax")))
                bad_file = True
            if bad_file:
                bad_labels.append(file_name)
    return bad_labels

def remove_file(bad_labels):
    for x in bad_labels:
        print("Removing file: {}".format(x))
```
