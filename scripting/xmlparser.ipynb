```python
# iterate through all xml files in D:\coding\EfficientDet\data\datasets\v1\Labels

import os
import xml.etree.ElementTree as ET


path = "D:/coding/EfficientDet/data/datasets/v1/Labels"
labels = {"metal", "glass", "paper", "cardboard", "pet", "hdpe", "pvc", "ldpe", "pp", "ps", "other"}


for file in os.listdir(path):

        base_name = os.path.splitext(file)[0]
        # get the path to the xml file

        file_name = os.path.join(path, base_name + ".xml")
        root = ET.parse(file_name).getroot()

        # read the <object> <name> tags
        # if the name does not exist in D:\coding\EfficientDet\data\datasets\v1/labels.txt
        # print the file name

        object = root.findall("object")

        for o in object:
            name = o.find("name")
            #print(name.text)

            if name.text not in labels:
                print("Bad label: {}".format(name.text))
                print("File name: {}".format(file_name))
                print("")
            else:
                print("Good label: {}".format(name.text))
                print("File name: {}".format(file_name))
                print("")





    

```
